```python
import bs4 as bs
import urllib.request
import csv

page = 1
alltitle = [['id', 'book_id', 'authors', 'title', 'average_rating', 'ratings_count']]

while page != 9:

    url = 'https://www.goodreads.com/list/show/567.Best_Chapter_Books_to_Read_Out_Loud?page=' + str(++page)

    grURL = urllib.request.urlopen(url)
    grHTML = grURL.read()
    grURL.close()

    soup = bs.BeautifulSoup(grHTML, 'html.parser')
    #grAllSpans = soup.find_all("span")
    spans = soup.find_all('a', title= True)
    for links in spans:
        alltitle.append(links['id'], links['book_id'], links['authors'], links['title'], links['average_rating'], links['ratings_count'])
                   
        myFile = open('books.csv', 'w')
        with myFile:
            writer = csv.writer(alltitle)
            writer.writerows(alltitle)
            
    if page == 9:
        break
    page += 1
```
